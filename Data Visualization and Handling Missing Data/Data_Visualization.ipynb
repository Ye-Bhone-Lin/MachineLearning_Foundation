### Matplotlib

Lineplot

```python
import matplotlib.pyplot as plt

y = [100, 1000, 1500, 2000, 4000]
x = [2020, 2021, 2022, 2023,2024]

plt.plot(x, y, marker='o')
plt.title("Inflation Rate")
plt.ylabel("Dollar Rate")
plt.xlabel("Year")
plt.show()
```

Histogram

```python
import numpy as np
import matplotlib.pyplot as plt
data = np.random.randn(1000)
#print(data)
plt.hist(data, bins=50, color='skyblue', edgecolor='black')
plt.title("Histogram of Normally Distributed Data")
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.show()
```

### Seaborn

Pairplot

```python
import seaborn as sns
iris = sns.load_dataset("iris")
print(iris)
sns.pairplot(iris, hue="species")
plt.show()
```

Heatmap

```python
iris = iris[['sepal_length','sepal_width',"petal_length","petal_width"]]
correlation = iris.corr()
print(correlation)
sns.heatmap(correlation, annot=True, cmap="coolwarm")
plt.title("Correlation Heatmap")
plt.show()
```

Boxplot

```python
import seaborn as sns
iris = sns.load_dataset("iris")
sns.boxplot(x="species", y="sepal_width", data=iris)
plt.title("Box Plot of Sepal Length by Species")
plt.show()
```

### Example EDA


```python
data = sns.load_dataset('titanic')
data.head()
```

```python
data.info()
```

```python
data.isnull().sum()
```

```python
data = data.dropna()
```

```python
data['age'] = data['age'].astype("int64")
```

```python
data
```

```python
sns.barplot(x="class", y="survived", data=data)
plt.title("Survival Rate by Class")
plt.show()
```

```python
sns.histplot(data, x="age", hue="survived", multiple="stack")
plt.title("Age Distribution by Survival")
plt.show()
```
