```python
import pandas as pd
import numpy as np
```

### Mean Imputation

```python
df_copy = df.copy() # To handle the missing values by using mean
df_copy
```

```python
df_copy.isnull().mean() # First of all check how many NaN values in this dataset
# If the NaN values mean is more than 0.5, we can drop this column.
```

```python
df_copy.isnull().sum()
```

```python
df_copy['Age'] = df_copy['Age'].replace(np.NaN, df_copy['Age'].mean())
df_copy
```

```python
df_copy['Age'] = df_copy['Age'].astype('int64') # Change a datatype float to int because age values can not be float datatype
```

```python
df_copy
```

### Median Imputation

```python
df_copy2 = df.copy()
df_copy2
```

```python
df_copy2['Age'] = df_copy2['Age'].replace(np.NaN,df_copy2['Age'].median())
df_copy2
```

```python
df_copy2.isnull().sum()
```

### Mode Imputation

```python
df_copy3 = df.copy()
df_copy3
```

```python
import statistics
df_copy3['Age'] = df_copy3['Age'].replace(np.NaN,statistics.mode(df_copy3['Age']))
df_copy3
```

```python
df_copy3.isnull().sum()
```

### Linear Interpolation

```python
df_copy4 = df.copy()
df_copy4
```

```python
df_copy4['Age'] = df_copy4['Age'].interpolate(method='linear',limit_direction='forward',axis=0)
df_copy4
```

```python
df_copy4.isnull().sum()
```
