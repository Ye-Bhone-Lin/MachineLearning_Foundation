```python
import pandas as pd
df = pd.read_csv("/content/sample_data/california_housing_train.csv")
df
```

```python
df1 = pd.read_csv("/content/sample_data/california_housing_test.csv")
df1
```

```python
final_df = pd.concat([df,df1])
final_df
```

```python
final_df = final_df.reset_index()
final_df
```

```python
final_df = final_df.drop(columns=['level_0', "index"])
final_df
```

```python
x = final_df.iloc[:,:-1]
x
```

```python
y = final_df.iloc[:,-1]
y
```

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.3, random_state=0)
```

```python
print(x_train.shape)
print(y_train.shape)
print(x_test.shape)
print(y_test.shape)
```

```python
from sklearn.svm import SVR

model = SVR()
model.fit(x_train,y_train)
```

```python
y_pred = model.predict(x_test)
```

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error

mean_squared_error(y_test,y_pred)
```

```python
mean_absolute_error(y_test,y_pred)
```

```python
from sklearn.model_selection import GridSearchCV

param_grid = {'C': [0.1, 1],
			'gamma': [1, 0.1],
			'kernel': ['rbf']}
```

```python
grid_search = GridSearchCV(model, param_grid=param_grid,refit = True, verbose = 3 )
```

```python
grid_search.fit(x_train,y_train)
```

```python
print(grid_search.best_params_)
print(grid_search.best_score_)
```

```python
latest_y_predict = grid_search.predict(x_test)
mean_squared_error(y_test,latest_y_predict)
```
