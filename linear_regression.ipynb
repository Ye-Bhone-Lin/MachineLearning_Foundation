```python
import pandas as pd 
import os 

print(os.getcwd())

train_df = pd.read_csv(os.getcwd()+"\california_housing_train.csv")
test_df = pd.read_csv(os.getcwd()+"\california_housing_test.csv")
train_df
```

```python
print(train_df.shape)
print(test_df.shape)
```

```python
drop_cols_train = train_df.drop(columns=['latitude','longitude'])
drop_cols_test = test_df.drop(columns=['latitude','longitude'])
```

```python
drop_cols_train
```

```python
train_x = drop_cols_train.iloc[:,:-1] # index location
train_y = drop_cols_train[['median_house_value']]
test_x = drop_cols_test.iloc[:,:-1]
test_y = drop_cols_test[['median_house_value']]
```

```python
from sklearn.preprocessing import StandardScaler

ss = StandardScaler()
train_x = ss.fit_transform(train_x)
train_y = ss.fit_transform(train_y)
test_x = ss.fit_transform(test_x)
test_y = ss.fit_transform(test_y)
```

```python
train_x
```

```python
from sklearn.linear_model import LinearRegression

model = LinearRegression().fit(train_x,train_y)
```

```python
y_predict = model.predict(test_x)
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error

mean_squared_error(test_y,y_predict)
```

```python
mean_absolute_error(test_y,y_predict)
```

```python
summ = sum([(y-y_predict)**2 for y,y_predict in zip(test_y,y_predict)])
```
